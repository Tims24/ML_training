```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
movie = pd.read_csv('movie_metadata.csv')
```

```python
movie.head()
```

```python
genres = movie[['movie_title', 'genres']]
```

```python
genres.head()
```

```python
def reverse(value):
    return value[::-1]
```

```python
for i, row in genres.iterrows():
    print(row)
```

```python
for i, row in genres.iteritems():
    print(row.map(reverse))
```
